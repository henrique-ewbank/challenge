```python
NFeNumber =[]
ProductName = []
Value = []
Quantity = []
i = 0
for i in range(len(dados)): #para cada linha i
    listao = list(dados.ItemList.iloc[i])
    j=0
    for j in range(len(listao)): #para cada item j da lista da linha i
        NFeNumber.append(dados['NFeNumber'].iloc[i])
        ProductName.append(dict(listao[j])['ProductName'])
        Value.append(dict(listao[j])['Value'])
        Quantity.append(dict(listao[j])['Quantity'])
ProductName
NFeNumber
ItemList_relacional = pd.DataFrame(list(zip(NFeNumber, ProductName, Value, Quantity)),
                                   columns=['NFeNumber','ProductName', 'Value', 'Quantity'])
```
